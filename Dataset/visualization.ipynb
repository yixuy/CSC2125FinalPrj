```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python

def read_csv_files(folder_path):
    # Pattern to match all CSV files in the folder
    pattern = f"{folder_path}/*.csv"
    csv_files = glob.glob(pattern)
    # Loop through the list of csv files
    plt.figure(figsize=(35, 20))
    for file in csv_files:
        # Read the current CSV file into a DataFrame
        df = pd.read_csv(file)
        temp = pd.to_datetime(df['datetime'])
        plt.plot( temp.dt.tz_convert(None), df['price'], label = file)
    plt.xticks(rotation=45)
    plt.grid(True)
    # Display the plot
    plt.tight_layout()
    plt.legend(loc="upper left")
    plt.show()
```

```python
folder_path = '../Dataset/doge_data'
read_csv_files(folder_path=folder_path)
```
